```python
import math
import random as rd
import matplotlib.pyplot as plt
import numpy as np
rds=[]
for i in range(1,1000):
    
    s=rd.randrange(1,1000)
    rds.append(s)
plt.figure(figsize=(20,5))
plt.plot(rds,c='red')
plt.title('noise')

sin=[]
for i in range(1,1000):
    s=math.sin(i)
    sin.append(s)

plt.figure(figsize=(20,5))
plt.title('sin signal')
plt.plot(sin,c='green')

sn=[]
tim=[]
fsn=[]
freq=[]
value=[]
for i in range(1,999):
    tim.append(i)
    f=(1/i)
    freq.append(f)
    snr=(sin[i]/rds[i])
    sn.append(snr)
    snrt=20*(np.log(sin[i]/rds[i]))
    value.append(snrt)


plt.figure(figsize=(20,5))
plt.plot(tim,sn,c='red')
plt.title('SNR')
plt.xlabel('time')
plt.ylabel('SNR')

plt.figure(figsize=(20,5))
plt.plot(tim,value,c='red')
plt.title('SNR')
plt.xlabel('time')
plt.ylabel('SNR')


w=np.fft.fft(sn)
plt.figure(figsize=(20,5))
plt.plot(freq,w.real,'g-',freq,w.imag,'r--')
plt.title('FFT vs Frequency')
plt.legend(('real', 'imaginary'))


wo=np.fft.ifft(sn)
plt.figure(figsize=(20,5))
plt.plot(tim,wo.real,'g-',tim,wo.imag,'r--')
plt.title('iFFT vs Time')
plt.legend(('real', 'imaginary'))
plt.show()

plt.figure(figsize=(20,5))
plt.specgram(wo, NFFT=512, Fs=fs,cmap='gist_rainbow_r')
plt.show()
```
